```python
from tensorflow.keras.layers import Embedding     # fpr word embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences   # for Padding
from tensorflow.keras.models import Sequential    # for creating the sequential model layers
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot  # helps us to do one hot encodding
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(dff)):
    review = re.sub('[^a-zA-Z]', " ", dff['text'][i] or " ")
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
corpus[1]
```

```python
df2['text'][0]
```

```python
# one Hot representation
voc_size = 5000
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
onehot_repr[1]
```

```python
sent_length=35
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)  # to make sure that every sentence should have same length
print(embedded_docs)
```

```python
embedded_docs[9]
```

```python
from tensorflow.keras.layers import Embedding  # like word2vec implementaion
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
## Creating model
embedding_vector_features=40 ##features representation # each and every word now going to convert in vector of 40 size,40 feature it may have
model=Sequential()     # each word represent 40 diffrent features
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))   # LSTM with 100 neurons   
model.add(Dense(1,activation='sigmoid'))   #
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())  # when ever my output is binary we use binary crossentropy, optimaization
```

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```
